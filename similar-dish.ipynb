```python
def stop_words(t):
    text = t.lower()
    stop_words = set(stopwords.words('english'))   
    word_tokens = word_tokenize(text)   
    filtered_sentence = [] 
    for w in word_tokens: 
        if w not in stop_words: 
            if w not in stop_food:
                filtered_sentence.append(w) 
            
    return filtered_sentence
```

```python
def tokenize(body):
    word_count = {}
    word_count = dict.fromkeys(food_tags, 0)
    stop_word = stop_words(body)
    for i in range(0, len(stop_word)):
        if stop_word[i] in veggie:
            word_count['veggie'] += 1            
        elif stop_word[i] in soup:
            word_count['soup'] += 1
        elif stop_word[i] in seafood:
            word_count['seafood'] += 1
        elif stop_word[i] in traditional:
            word_count['traditional'] += 1
        elif stop_word[i] in street:
            word_count['street'] += 1
        elif stop_word[i] in sweet:
            word_count['sweet'] += 1
    return word_count
```

```python
def add_title(df):
    bag_of_word['food'] = df['title']
    for i in range(0, len(df)):
        word_count_dict = tokenize(df['body'][i])
        for k,e in word_count_dict.items():
            bag_of_word[k][i] = e
```
